```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cleancsv(df):
    df = df.copy()
    df = df.drop(df.columns[0],axis=1)
    col = ["br" +str(i) for i in range (1, 26)] + ["in" +str(i) for i in range (1, 26)] + ["ba" +str(i) for i in range (1, 26)]  + ["hf" +str(i) for i in range (1, 26)] 
    df.columns = col
    df = df.apply(pd.to_numeric)
    df.index = col
    return df
```

```python
df = pd.read_csv('94graph.csv')
df = cleancsv(df)
new = pd.read_csv('graph06.csv')
new = cleancsv(new)
```

```python
sum(np.sum(new<0.05))

```

```python
sum(np.sum(df<0.05))
```

```python
frame= df
res = []

for i in range(100):
    for j in range(100):
        if(i!=j):
            res.append((frame.iloc[i,j], frame.iloc[j,i]))
```

```python
frame= new
newres = []

for i in range(100):
    for j in range(100):
        if(i!=j):
            newres.append((frame.iloc[i,j], frame.iloc[j,i]))
```

```python
plt.scatter([p[0] for p in res], [p[1] for p in res])
plt.title('old res');
```

```python
plt.scatter([p[0] for p in newres], [p[1] for p in newres])
plt.title('new res');
```

```python
plt.hexbin([p[0] for p in res], [p[1] for p in res], gridsize=10)
plt.colorbar()
plt.title("for 1994");
```

```python

plt.hexbin([p[0] for p in newres], [p[1] for p in newres], gridsize=10)
plt.colorbar()
plt.title("for 2006");
```
